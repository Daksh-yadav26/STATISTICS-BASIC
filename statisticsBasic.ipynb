Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Answer- Descriptive Statistics

Definition: Descriptive statistics summarize and describe the main features of a dataset.

Purpose: To present data in a meaningful way using numbers, tables, or graphs.

Examples:

Mean, median, mode (measures of central tendency)

Range, variance, standard deviation (measures of dispersion)

Frequency distributions, histograms

Example:
If a teacher records the test scores of 50 students and calculates the average score (mean = 75), that’s descriptive statistics — it only describes the data collected.

Inferential Statistics

Definition: Inferential statistics use a sample of data to make generalizations or predictions about a larger population.

Purpose: To draw conclusions beyond the immediate data.

Examples:

Hypothesis testing

Confidence intervals

Regression analysis

Example:
A researcher surveys 100 voters to estimate how all voters in a country will vote. The sample data are used to infer characteristics of the entire population.

Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.

Answer- Sampling is the process of selecting a subset (sample) from a population to gather data and make inferences about the entire population, especially when studying everyone is impractical.

Random Sampling

Definition: Every individual in the population has an equal chance of being selected.

Example: Picking 50 names randomly from a list of 1,000 employees.

Advantage: Reduces selection bias; sample is likely to be representative of the population.

Stratified Sampling

Definition: The population is divided into subgroups (strata) based on shared characteristics (like age, gender, income), and random samples are taken from each stratum.

Example: Dividing students by grade level and then randomly selecting a few from each grade.

Advantage: Ensures representation of all important subgroups, increasing accuracy.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?

Answer- Skewness

Definition: Skewness measures the asymmetry of a data distribution.

Positive Skew (Right Skew): The right tail is longer — most data values are on the lower end, but a few large values stretch the tail rightward.

Example: Income data — most people earn moderate salaries, but a few earn very high ones.

Negative Skew (Left Skew): The left tail is longer — most data are high, but a few small values pull the mean leftward.

Kurtosis

Definition: Kurtosis measures the “peakedness” or tailedness of a distribution.

High Kurtosis (Leptokurtic): Sharp peak, heavy tails → more outliers.

Low Kurtosis (Platykurtic): Flatter peak, lighter tails → fewer outliers.

Normal Kurtosis (Mesokurtic): Similar to the normal distribution.

Interpretation of Positive Skew

A positive skew means:

The tail extends to the right.

The mean > median > mode.

The dataset contains a few very large values that increase the average.

```python
#Question 6: Compute the covariance and correlation coefficient between the followingtwo datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]
#Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

import numpy as np

# Compute means
mean_x = np.mean(list_x)
mean_y = np.mean(list_y)

# Compute covariance manually
covariance = np.mean([(x - mean_x) * (y - mean_y) for x, y in zip(list_x, list_y)])

# Compute correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

# Display results
print("Mean of X:", mean_x)
print("Mean of Y:", mean_y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list andidentify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Determine outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display results
print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.

Answer-. Covariance

Meaning: Covariance measures how two variables change together.

If both increase together → positive covariance

If one increases while the other decreases → negative covariance

If the value is close to zero → little to no relationship

 Interpretation example:
If covariance > 0, higher ad spending tends to coincide with higher sales

Correlation Coefficient

Meaning: Correlation is the standardized form of covariance — it measures both the direction and strength of the linear relationship.

Range:

+1 → Perfect positive relationship

0 → No linear relationship

-1 → Perfect negative relationship

 Interpretation example:
A correlation close to +1 means that when advertising spend increases, daily sales increase strongly.

```python
#Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales
#Write Python code to compute the correlation between the two lists:
#advertising_spend = [200, 250, 300, 400, 500]
#daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance matrix
cov_matrix = np.cov(advertising_spend, daily_sales, bias=True)
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.

Answer-Summary Statistics to Use

1-Mean (Average)

The average satisfaction score

Shows overall customer satisfaction level

2-Median

The middle score when sorted

Useful if the data is skewed or has outliers

3-Mode

   The most frequent rating

   Shows the most common customer opinion

4-Standard Deviation (SD)


  How spread out the scores are

  Indicates consistency of satisfaction low SD means most customers agree, high SD means mixed opinions

5-Minimum & Maximum

Lowest and highest scores

Show the range of customer satisfaction

Visualization Techniques

1-Histogram

Displays how many customers gave each rating — helps see the distribution shape (e.g., normal, skewed)

2-Boxplot

Identifies the median, quartiles, and possible outliers

3-Bar Chart (for mode)

Shows the frequency of each score visually



```python
#Question 9: Your team has collected customer satisfaction survey data on a scale of
#1-10 and wants to understand its distribution before launching a new product.
#Write Python code to create a histogram using Matplotlib for the survey data:
#survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
import matplotlib.pyplot as plt
import numpy as np
import statistics

# Given data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Compute summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
mode_score = statistics.mode(survey_scores)
std_dev = np.std(survey_scores)

# Display summary
print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", std_dev)

# Create histogram
plt.hist(survey_scores, bins=range(1, 12), color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
